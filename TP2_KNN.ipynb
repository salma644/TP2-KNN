```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split

df=pd.read_csv("advertising.csv")

df.isnull().sum()
x=df[["TV","Radio","Newspaper"]]
y=df["Sales"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_test.head()

model=LinearRegression()

#apprentissage
model.fit(x_train,y_train)

#test
y_predict=model.predict(x_test)

#calcul de l'erreur quadratique(il faut quil soit mini,um pour que la prediction soit bonne et proche a la realité)
mse=mean_squared_error(y_test, y_predict)

#affichage
print("Erreur quadratique moyenne MSE",mse)
print("Coefficient a",model.coef_)
print("Coefficient b",model.intercept_)

plt.scatter(x_test["TV"], y_predict, color="black")

plt.plot(x_test["TV"], y_predict, color='green', linewidth=3)

plt.xlabel("TV")
plt.ylabel("Sales")
```
