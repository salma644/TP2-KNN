```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("advertising.csv")
```

```python
df.isnull().sum()
x=df[["TV","Radio","Newspaper"]]
y=df["Sales"]
```

```python

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_test.head()
```

```python
model=LinearRegression()
```

```python
#apprentissage
model.fit(x_train,y_train)
```

```python

#test
y_predict=model.predict(x_test)
```

```python
#calcul de l'erreur quadratique(il faut quil soit mini,um pour que la prediction soit bonne et proche a la realité)
mse=mean_squared_error(y_test, y_predict)
```

```python
#affichage
print("Erreur quadratique moyenne MSE",mse)
print("Coefficient a",model.coef_)
print("Coefficient b",model.intercept_)
```

```python
plt.plot(x_test, y_predict,color='Green',linewidth=3)
```
